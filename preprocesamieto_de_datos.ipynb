## Fusión de las bases de datos  (Esto es mejor por SQL. Primero hacer la validacion de campos a imputar, variables a eliminar)

Vamos a unir las bases de datos en una sola para facilidad de manejo

```python
# Extraer el año de las columnas de fecha y agregar la columna Year
df_general['Year'] = pd.to_datetime(df_general['InfoDate']).dt.year
df_encuesta_empleado['Year'] = pd.to_datetime(df_encuesta_empleado['DateSurvey']).dt.year
df_encuesta_gerente['Year'] = pd.to_datetime(df_encuesta_gerente['SurveyDate']).dt.year

#  Asegurarse de que EmployeeID y Year sean del mismo tipo en todos los DataFrames
df_general['EmployeeID'] = df_general['EmployeeID'].astype(int)
df_general['Year'] = df_general['Year'].astype(int)

df_encuesta_empleado['EmployeeID'] = df_encuesta_empleado['EmployeeID'].astype(int)
df_encuesta_empleado['Year'] = df_encuesta_empleado['Year'].astype(int)

df_encuesta_gerente['EmployeeID'] = df_encuesta_gerente['EmployeeID'].astype(int)
df_encuesta_gerente['Year'] = df_encuesta_gerente['Year'].astype(int)

# Fusionar df_general, df_encuesta_empleado y df_encuesta_gerente
df_completo_sin_retiro = pd.merge(df_general, df_encuesta_empleado, on=['EmployeeID', 'Year'], how='left')
df_completo_sin_retiro = pd.merge(df_completo_sin_retiro, df_encuesta_gerente, on=['EmployeeID', 'Year'], how='left')

# Fusionar df_completo_sin_retiro con df_retiro usando EmployeeID como clave
df_completo = pd.merge(df_completo_sin_retiro, df_retiro, on='EmployeeID', how='left')

# Mostrar los valores nulos por columna
# Calcular la cantidad de valores nulos por columna
nulos_por_columna = df_completo.isnull().sum()

# Mostrar los valores nulos en formato de tabla con todas las columnas
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

print("Valores nulos por columna en df_completo:")
print(nulos_por_columna)
```

```python
# Obtener los EmployeeID que tienen datos en 2015
ids_2015 = df_completo[df_completo['Year'] == 2015]['EmployeeID'].unique()

# Obtener los EmployeeID que tienen datos en 2016
ids_2016 = df_completo[df_completo['Year'] == 2016]['EmployeeID'].unique()

# Verificar cuántos EmployeeID tienen datos en ambos años
ids_en_ambos_anios = set(ids_2015).intersection(set(ids_2016))

# Verificar cuántos EmployeeID tienen datos solo en 2015 o solo en 2016
ids_solo_en_2015 = set(ids_2015) - set(ids_2016)
ids_solo_en_2016 = set(ids_2016) - set(ids_2015)

print(f"Cantidad de EmployeeID con datos en ambos años: {len(ids_en_ambos_anios)}")
print(f"Cantidad de EmployeeID con datos solo en 2015: {len(ids_solo_en_2015)}")
print(f"Cantidad de EmployeeID con datos solo en 2016: {len(ids_solo_en_2016)}")

# Ver algunos ejemplos de EmployeeID que tienen datos en ambos años
print(f"Ejemplos de EmployeeID con datos en ambos años: {list(ids_en_ambos_anios)[:10]}")
print(f"Ejemplos de EmployeeID con datos solo en 2015: {list(ids_solo_en_2015)[:10]}")
print(f"Ejemplos de EmployeeID con datos solo en 2016: {list(ids_solo_en_2016)[:10]}")
```
