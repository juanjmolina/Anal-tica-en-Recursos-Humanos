```python
# Importar librerias necesarias
import pandas as pd
```

```python
# Carga de  bases de datos 
# employee_survey_data.csv:  Encuesta realizada a los empleados sobre satisfacción laboral a final de cada año, 
# se tienen el histórico de dos encuestas, la realizada el 31-12-2015 y la que se realizó el 31-12-2016.
df_encuesta_empleado = pd.read_csv('https://raw.githubusercontent.com/osiris-contreras/Anal-tica-en-Recursos-Humanos/main/bases%20de%20datos%20RRHH/employee_survey_data.csv')

#general_data.csv: Información general de los empleados, dado que la información puede cambiar con el tiempo
#  se extrajo la información con el último corte 31-12-2016 y un histórico con la información con corte 31-12-2015
df_general = pd.read_csv('https://raw.githubusercontent.com/osiris-contreras/Anal-tica-en-Recursos-Humanos/main/bases%20de%20datos%20RRHH/general_data.csv')

# manager_survey_data.csv: Encuesta de desempeño de los empleados realizada por parte de los jefes, se tienen el histórico de dos encuestas,
#  la realizada el 31-12-2015 y la que se realizó el 31-12-2016.
df_encuesta_gerente = pd.read_csv('https://raw.githubusercontent.com/osiris-contreras/Anal-tica-en-Recursos-Humanos/main/bases%20de%20datos%20RRHH/manager_survey.csv')

# retirement_info.csv: Información de retiro de los empleados que dejaron la empresa, se tiene la información de retiro de los empleados 
# de 2015 y 2016, sin embargo, la información de 2015 no se pudo obtener completa por problemas en el almacenamiento de la información, 
# por lo tanto, no están todos los empleados que se retiraron ese año.
df_retiro = pd.read_csv('https://raw.githubusercontent.com/osiris-contreras/Anal-tica-en-Recursos-Humanos/main/bases%20de%20datos%20RRHH/retirement_info.csv')
```

```python
# Verificar la cantidad de valores nulos en cada DataFrame
print("\nValores nulos en df_general:")
print(df_general.isnull().sum())
```

```python
# Obtener el valor mínimo y máximo de las variables con datos nulos en df_general
print("Valores de NumCompaniesWorked:")
print("Mínimo:", df_general['NumCompaniesWorked'].min())
print("Máximo:", df_general['NumCompaniesWorked'].max())

print("\nValores de TotalWorkingYears:")
print("Mínimo:", df_general['TotalWorkingYears'].min())
print("Máximo:", df_general['TotalWorkingYears'].max())
```

df_general:
* NumCompaniesWorked (38 valores nulos): Dado que esta variable indica el número de empresas en las que un empleado ha trabajado, se pueden imputar estos valores nulos con la mediana, o incluso con cero si consideramos que estos empleados podrían haber trabajado solo en la empresa actual, esto pues si el cero no estaba presenta en la escala de datos, pero como si lo está, se imputaran con la mediana.
* TotalWorkingYears (18 valores nulos): Similar a NumCompaniesWorked, se pueden imputar estos valores con la mediana. 

```python
# Imputación de valores nulos en df_general utilizando la media
df_general['NumCompaniesWorked'].fillna(df_general['NumCompaniesWorked'].median(), inplace=True)
df_general['TotalWorkingYears'].fillna(df_general['TotalWorkingYears'].median(), inplace=True)
```

```python
print("Valores nulos en df_encuesta_empleado:")
print(df_encuesta_empleado.isnull().sum())

print("\nValores nulos en df_encuesta_gerente:")
print(df_encuesta_gerente.isnull().sum())

print("\nValores nulos en df_retiro:")
print(df_retiro.isnull().sum())
```

df_encuesta_empleado:
* EnvironmentSatisfaction (50 valores nulos): Esta es una variable importante, por lo que vamos a usar la  mediana.
* JobSatisfaction (40 valores nulos): Al igual que la anterior, esta variable también es crítica. Usaremos la mediana.
* WorkLifeBalance (76 valores nulos): Imputar con la mediana es una buena opción aquí también, dado que es una variable relevante para analizar la rotación.

df_encuesta_gerente:
No tiene valores nulos

df_retiro:
* resignationReason (70 valores nulos): Dado que esta variable es el motivo del retiro del empleado,  y los valores nulos coinciden con los empleados que fueron despedidos por la empresa, se puede crear una nueva categoría, como "Fired", para este caso.

```python
# Imputación de valores nulos en df_encuesta_empleado utilizando la mediana
df_encuesta_empleado['EnvironmentSatisfaction'].fillna(df_encuesta_empleado['EnvironmentSatisfaction'].median(), inplace=True)
df_encuesta_empleado['JobSatisfaction'].fillna(df_encuesta_empleado['JobSatisfaction'].median(), inplace=True)
df_encuesta_empleado['WorkLifeBalance'].fillna(df_encuesta_empleado['WorkLifeBalance'].median(), inplace=True)

# Imputación de valores nulos en df_retiro con la categoría "Fired"
df_retiro['resignationReason'].fillna('Fired', inplace=True)
```
